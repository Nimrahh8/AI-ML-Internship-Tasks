Task 1: Exploring and Visualizing the Iris Dataset

This notebook explores the famous Iris dataset, performs basic data analysis, visualizes features, and examines relationships between flower species and features.

Step 1: Import Libraries
We import the required libraries for data manipulation and visualization:
- `pandas` for handling tabular data
- `matplotlib.pyplot` and `seaborn` for creating plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Load Dataset

We use Seaborn's built-in Iris dataset, which contains 150 samples of iris flowers with features:
- sepal_length
- sepal_width
- petal_length
- petal_width
- species (target)

```python
df = sns.load_dataset("iris")

# Display first 5 rows
df.head()
```

Step 3: Explore Dataset

We check the structure, column names, and summary statistics of the dataset.

```python
print("Shape of dataset:", df.shape)
```

Step 4: Visualizations

Visualizations help understand relationships between features and detect patterns in the data.

```python
# Scatter Plot: Sepal Length vs Petal Length
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
# Histogram of all numeric features
df.hist(figsize=(8,6))
plt.tight_layout()
plt.show()
```

```python
# Boxplot of all features
plt.figure(figsize=(8,5))
sns.boxplot(data=df)
plt.title("Boxplot of Iris Features")
plt.show()
```

Step 5: Insights

- The scatter plot shows that petal length separates species better than sepal length.
- Histograms indicate numeric features are roughly normally distributed.
- Boxplots help identify feature ranges and potential outliers.
- Overall, the dataset is clean and suitable for classification tasks.
